```python
from pathlib import Path
import pandas as pd

in_path  = Path(r'../data/processed/final_cleaned_jokes.csv')
out_path = Path(r'../data/processed/final_cleaned_jokes_3topics.csv')

in_path, out_path
```

```python
df = pd.read_csv(in_path, dtype=str).fillna("")
df.head()
df.columns
```

```python
def count_topic_terms(topic_str: str) -> int:
    # Ensure it's a string
    s = str(topic_str)
    # Split by comma and strip spaces
    parts = [p.strip() for p in s.split(",") if p.strip()]
    return len(parts)

df["topic_count"] = df["topic"].apply(count_topic_terms)
df[["topic", "topic_count"]].head()

```

```python
df_3 = df[df["topic_count"] == 3].copy()
print("Full dataset:", df.shape)
print("3-topic subset:", df_3.shape)
```

```python
df_3 = df_3.drop(columns=["topic_count"])
```

```python
df_3.to_csv(out_path, index=False)
print("Saved 3-topic subset to:", out_path)
```
